```python
import pandas as pd
import numpy as np
from datetime import date
import calendar
```

```python
df = pd.read_excel('customerdata.xls', sheet_name='customerdata')

# print whole sheet data
print(df)
```

```python
df.head()
```

```python
df['transaction_days']=round((df['Day of Last Transaction']-df['Day of First Transaction'])/ np.timedelta64(1, 'D'))
```

```python
df.head()
```

```python
df['transaction_months']=round((df['Day of Last Transaction']-df['Day of First Transaction'])/ np.timedelta64(1, 'M'))
```

```python
df['transaction_weeks']=round((df['Day of Last Transaction']-df['Day of First Transaction'])/ np.timedelta64(1, 'W'))
```

```python
df['transaction_Years']=round((df['Day of Last Transaction']-df['Day of First Transaction'])/ np.timedelta64(1, 'Y'))
```

```python
df.head() 
```

```python
df['first_transaction_after_approval']=round((df['Day of First Transaction']-df['Day of Preapproved On'])/ np.timedelta64(1, 'D'))
```

```python
df['Day of First Transaction weekday']=df['Day of First Transaction'].dt.day_name()
```

```python
df['Day of Last Transaction weekday']=df['Day of Last Transaction'].dt.day_name()
```

```python
df['Day of Preapproved On weekday']=df['Day of Preapproved On'].dt.day_name()
```

```python
df
```

```python
df['Day of First Transaction month']=pd.DatetimeIndex(df['Day of First Transaction']).month
df['Day of Last Transaction month']=pd.DatetimeIndex(df['Day of Last Transaction']).month
df['Day of Preapproved On month'] = pd.DatetimeIndex(df['Day of Preapproved On']).month
                                                     
```

```python
df
```

```python
df['Average daily volume processed'] = round(df['Total Volume']/df['transaction_days'])
df['Average monthly volume processed'] = round(df['Total Volume']/df['transaction_months'])
df['Average yearly volume processed'] = round(df['Total Volume']/df['transaction_Years'])
df['Average weekly volume processed'] = round(df['Total Volume']/df['transaction_weeks'])

```
